## Task 11. For each user ‘a’ and ‘b’, calculate the count of rows where the gt column contains ‘stairs’ and Device = ‘nexus4_1’

```
# Set up the transformation
stairsonlyDF = streamingDF.filter((streamingDF["Device"] == "nexus4_1") &
                                  (streamingDF["gt"].contains("stairs")) &
                                  (streamingDF["User"].isin(["a", "b"]))) \
    .groupBy("User") \
    .count()

```

```
# Set up the output sink
query_name_stairsonly = "stairsonlyDF"
query_stairsonly = stairsonlyDF.writeStream \
    .queryName(query_name_stairsonly) \
    .format("memory") \
    .outputMode("update") \
    .start()
```

```
# View the streaming output
for _ in range(5):
    spark.sql(f"SELECT * FROM {query_name_stairsonly}").show()
    sleep(5)
```

```
# Stop the query
query_stairsonly.stop()
```

## Task 12. Compute the averages of each column ‘x’, ‘y’, ‘z’ and counts of Device column for each combination of values for ‘Model’ and ‘gt’ columns

```
# Set up the transformations

from pyspark.sql.functions import *

gtmodelstatsDF = streamingDF.groupby("gt", "Model") \
    .agg(avg("x").alias("x_avg"), avg("y").alias("y_avg"), avg("z").alias("z_avg"), count("Device").alias("devicecount"))

```

```
# Set up the output sink
query_name_gtmodelstats = "gtmodelstatsQuery"
query_gtmodelstats = gtmodelstatsDF.writeStream \
    .queryName(query_name_gtmodelstats) \
    .format("memory") \
    .outputMode("complete") \
    .start()

```

```


# View the streaming output with a filter applied
for _ in range(5):
    spark.sql(f"SELECT * FROM {query_name_gtmodelstats} WHERE x_avg > 0").show()
    sleep(5)
```

```
# Stop the query
query_gtmodelstats.stop()
```

## Task 13. Perform a join of a static DataFrame with a streaming DataFrame

```
# Compute the historical averages
historicalAvgDF = staticDF.groupby("gt", "Model") \
    .agg(avg("x").alias("x_avg_h"), avg("y").alias("y_avg_h"), avg("z").alias("z_avg_h"))

```

```
# Calculate the averages for the streaming DataFrame
streamingAvgDF = streamingDF.groupby("gt", "Model") \
    .agg(avg("x").alias("x_avg_s"), avg("y").alias("y_avg_s"), avg("z").alias("z_avg_s"))

```

```
# Join the historical averages with the streaming averages
joinedDF = historicalAvgDF.join(streamingAvgDF, ["gt", "Model"])

```

```
# Set up the output sink
query_name_join = "joinQuery"
query_join = joinedDF.writeStream \
    .queryName(query_name_join) \
    .format("memory") \
    .outputMode("complete") \
    .start()

```

```
# Display the streaming output (the joined table) 5 times
for _ in range(5):
    spark.sql(f"SELECT * FROM {query_name_join}").show()
    sleep(5)
```

```
# Stop the query
query_join.stop()
```
